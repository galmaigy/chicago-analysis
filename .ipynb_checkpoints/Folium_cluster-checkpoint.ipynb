```python
import folium
import folium.plugins as plugins
from folium.plugins import HeatMapWithTime
from folium.plugins import HeatMap

df_map = df.copy()
df_map=df_map.dropna(subset=['latitude','longitude'])
df_map.columns = [x.lower() for x in df_map.columns]
df_map['count']=1

df_s = df_map.sample(10000)
```

```python
from folium.plugins import FastMarkerCluster
folium_map = folium.Map(location=[41.881832 , -87.623177],
                        zoom_start=11,
                        tiles='CartoDB dark_matter')

# These two lines should create FastMarkerClusters
FastMarkerCluster(data=list(zip(df_s['latitude'].values, df_s['longitude'].values))).add_to(folium_map)
folium.LayerControl().add_to(folium_map)


for index, row in df_s.iterrows():
    folium.CircleMarker(location=(row["latitude"],
                                  row["longitude"]),
                        radius= 5,
                        color="#007848",
#                         popup=popup_text,
                        fill=False).add_to(folium_map)
folium_map.save('map/cluster_10000.html')
```
